```r
#columB <- sup$Branch
#head(columB)
supubdate <- as.integer(sat[[25]])
supubdate<- ifelse( sat[25]== "neutral or dissatisfied", 0,1)
supubdate<-as.data.frame(supubdate)
head(supubdate)
print(class(supubdate))

# إضافة العمود الجديد إلى DataFrame باستخدام cbind()
data <- cbind(sat, supubdate)

# تسمية العمود الجديد
colnames(sat)[ncol(sat)] <- "new_column"

# طباعة DataFrame مع العمود الجديد
head(data)
```

```r
sate <- na.omit(sat)
colSums(is.na(sate))
```

```r
# Load the required libraries
library(caret)
library(rpart)

# Split the data into training and testing sets
set.seed(123)
trainIndex <- createDataPartition(data$satisfaction, p = .8, list = FALSE, times = 1)
data_train <- data[trainIndex, ]
data_test <- data[-trainIndex, ]

# Make sure factor levels are consistent
#data_test$Product.line <- factor(data_test$Product.line, levels = levels(data_train$Product.line))

# Train the logistic regression model
model <- train(satisfaction ~Food_and_drink+Cleanliness+Inflight_wifi_service+Ease_of_Online_booking, data = data_train, method = "rpart", trControl = trainControl(method = "cv"))

# Make predictions on the testing set
predictions <- predict(model, newdata = data_test)

# Evaluate the model

predictions = as.factor(predictions)
data_test$Product.line = as.factor( data_test$satisfaction)

#confusionMatrix(predictions, data_test$Product.line)

x <- data.frame(predictions, data_test$satisfaction)
print(x)
```

```r
# Calculate accuracy
accuracy <- sum(predictions == data_test$satisfaction) / length(data_test$satisfaction) * 100
cat("Accuracy:", accuracy,"%\n")
```

```r
# Calculate recall
conf_matrix <- table(predictions, data_test$satisfaction)
recall <- diag(conf_matrix) / rowSums(conf_matrix) * 100
cat("Recall:",recall,"%\n")
```

```r
# Calculate precision
conf_matrix <- table(predictions, data_test$satisfaction)
precision <- diag(conf_matrix) / colSums(conf_matrix) * 100
cat("Precision:", precision,"%\n")
```

```r
# Calculate F-measure
conf_matrix <- table(predictions, data_test$satisfaction)
precision <- diag(conf_matrix) / colSums(conf_matrix)
recall <- diag(conf_matrix) / rowSums(conf_matrix)
f_measure <- 2 * (precision * recall) / (precision + recall)
cat("F-measure:", f_measure,"\n")
```

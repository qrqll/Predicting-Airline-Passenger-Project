```r

flight_distance <- sat$Flight_distance
age <- sat$Age

# حساب المتوسط والانحراف المعياري
mean_distance <- mean(flight_distance)
sd_distance <- sd(flight_distance)

mean_age <- mean(age)
sd_age <- sd(age)

# تحديد الحد الأدنى للتصنيف كشذوذ
threshold_distance <- mean_distance - 3 * sd_distance
threshold_age <- mean_age + 2 * sd_age

# استخراج السجلات الشاذة
anomalies <- sat[(flight_distance < threshold_distance) | (age > threshold_age), ]

# طباعة السجلات الشاذة
print(anomalies)

# Create the plot
plot <- ggplot(sat, aes(x = Flight_distance, y = Age)) +
  geom_point() +  # Draw points
  stat_smooth(method = "lm", se = FALSE) +  # Draw a trend line
  labs(x = "Flight Distance", y = "Age")  # Set axis labels

# Print the plot
print(plot)
```

```r
library(ggplot2)
library(data.table)
#اعتمديه
# تحويل البيانات إلى كائن data.table لتحسين الأداء
sat <- as.data.table(sat)

Flight_Distance <- sat$Flight_Distance
Age <- sat$Age

# حساب المتوسط والانحراف المعياري
mean_distance <- mean(Flight_Distance)
sd_distance <- sd(Flight_Distance)

mean_Age <- mean(Age)
sd_Age <- sd(Age)

# تحديد الحد الأدنى للتصنيف كشذوذ
threshold_distance <- mean_distance - 3 * sd_distance
threshold_age <- mean_Age + 2 * sd_Age

# استخراج السجلات الشاذة باستخدام التعبيرات الشرطية في data.table
anomalies <- sat[Flight_Distance < threshold_distance | Age > threshold_age]

# طباعة السجلات الشاذة
anomalies %>% select(Flight_Distance,Age)
nrow(anomalies)
# إنشاء الرسم البياني
plot <- ggplot(sat, aes(x = Flight_Distance, y = Age)) +
  geom_point() +  # رسم النقاط
  geom_smooth(method = "lm", se = FALSE) +  # رسم خط اتجاه
  labs(x = "Flight Distance", y = "Age")  # تعيين تسميات المحاور

# طباعة الرسم البياني
print(plot)
```
